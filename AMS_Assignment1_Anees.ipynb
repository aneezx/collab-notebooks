```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import scipy.linalg
import sympy
```

```python
whitedf = pd.read_csv("winequality-white.csv", delimiter=';')
reddf = pd.read_csv("winequality-red.csv", delimiter=';')
```

```python
arrWhite = whitedf.to_numpy()
```

```python
arrRed = reddf.to_numpy()
```

##1. Vector and Matrix Algebra

```python
#Scalar Addition
5 + arrRed
```

```python
#Matrix multiplication cannot be performed as the shapes are different
#Calculating Rank for White wine
rankWhiteWine = np.linalg.matrix_rank(arrWhite)
print('Rank of White wine matrix is : ', rankWhiteWine)
```

```python
#Calculating Rank for Red wine
rankRedWine = np.linalg.matrix_rank(arrRed)
print('Rank of Red wine matrix is : ', rankRedWine)
```

##2. Diagonalization and Determinants

```python
#Generate the covariance matrix
covWhiteWine = np.cov(arrWhite)

covRedWine = np.cov(arrRed)
```

```python
#Generate the determinant of White wine
print('Determinant of covariant of White wine is ',np.linalg.det(covWhiteWine))

#Generate the determinant of Red wine
print('Determinant of covariant of Red wine is ',np.linalg.det(covRedWine))
```

```python
#Diagonalise White wine
print('Diagonal matrix of White wine is ',covWhiteWine.diagonal())

#Diagonalise Red wine
print('Diagonal matrix of Red wine is ', covRedWine.diagonal())
```

##3. Eigen Values and Eigen Vectors

Eigen vectors are the principal components. As generated below, the first column (1.67e-02+0.j, -1.45e-17+0.j, etc) is the first principal component and so on. Each vector can be scaled to a specific eigen value whose magnitude says how much the data variability is.

```python
#Generate the eigen value and eigen vector for white wine
wval, wvec = np.linalg.eig(covWhiteWine)

print('The eigen values of covariance of white wine is ', wval)

print('The eigen vectors of covariance of white wine is ', wvec)
```

```python
#Generate the eigen value and eigen vector for red wine
rval, rvec = np.linalg.eig(covRedWine)

print('The eigen values of covariance of red wine is ', rval)

print('The eigen vectors of covariance of red wine is ', rvec)
```

##4. Principal Component Analysis (PCA)
###Picking the white wine dataset

```python
#Calculate the explained variance

w_var_exp = []

for i in wval:
  w_var_exp.append((i / sum(wval)) * 100)

print(w_var_exp) #98% of variance in our data is explained by our first principal component, 0.85 by second and so on

cum_w_var_exp = np.cumsum(w_var_exp)

print(cum_w_var_exp)
```

```python
#Visualisation
sns.lineplot(x = range(4898), y=cum_w_var_exp)
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.title("Explained variance vs Number of components")
```

```python
#Project the data into lower dimension
projection_matrix = (wvec.T[:][:2]).T
print(projection_matrix)

PCA = covWhiteWine.dot(projection_matrix)
```

##4. Single Value Decomposition

This provides you the lower dimensional approximation of a matrix which helps us to reduce the less important data.

```python
#Compute the factor by SVD
U, s, V = np.linalg.svd(arrWhite, full_matrices = False)

print('Factor of the given array by SVD\n')
print('U = ', U, '\n\ns = ', s, '\n\nV = ', V)
```

## 5. LU Decomposition and Gaussian Elimination

LU decomposition is the best way to acheive Gaussian elimination as this would decompose the matrix into upper and lower triangular matrices through row level operations

```python
P, L, U = scipy.linalg.lu(arrWhite)

print('LU Decomposition results\n')
print('P:', P, '\n\nL:', L, '\n\nU:', U)
```

##6. QR Factorization

QR factorization of a matrix is the decomposition of a matrix say A into A=QR where Q is orthogonal and R is an upper-triangular matrix

```python
q, r = np.linalg.qr(arrWhite)

print('\nQ:\n', q)
print('\nR:\n', r)
```

##10. Linear dependence and Linear Independence



```python
matrix = np.array([
  [0, 5],
  [0, -10]
])

_, indexes = sympy.Matrix(matrix).T.rref()  # T is for transpose
print(indexes) #Indices of the linearly independent rows

print('\n\n',matrix[indexes,:]) #Linearly independent matrix

if np.linalg.det(matrix) == 0:
    print("\n\nlinearly dependant")
else:
    print("\n\nlinearly independant")
```
