```python
class Person:
  def __init__(self, age, name, weight): #constructor
    self.name = name
    self.age = age
    self.weight = weight

p1 = Person(39,'John',90)
p2 = Person(40,'Harold',93)

print(p1.name)
print(p2.name)
```

```python
#Simple Inheritance
class Animal:
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def eat(self):
    print(self.name, 'is eating')

  def specs(self):
    print("The dog's name is ",self)

class Dog(Animal):
  def __init__(self, name, color, breed):
    Animal.__init__(self, name, color)
    self.breed = breed
```

```python
doggie = Dog('Spike','White','Dobberman')
doggie.eat()
#doggie.specs()
```
