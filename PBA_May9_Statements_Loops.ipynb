```python
num=int(input('enter your num: '))
if num % 2 == 0:
    if num > 100:
        print("The Num is even and greater than 100 ")
    else:
        print("The num is even but less than 100 ")
else:
    print("The num is odd")
```

```python
num=int(input('enter your num: '))
if num >= 0:
    if num == 0:
        print("The Num is zero")
    else:
        print("The num is positive")
else:
    print("The num is negative")
```
