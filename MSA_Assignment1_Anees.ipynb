```python
# Hypothesis testing
# Define the null and alternative hypotheses
null_hypothesis = "\nThere is no significant difference between the mean age of customers who subscribed to the term deposit and those who did not."
alternative_hypothesis = "\nThere is a significant difference between the mean age of customers who subscribed to the term deposit and those who did not."

# Extract the age of customers who subscribed and did not subscribe
subscribed = df[df['y'] == 'yes']['age']
not_subscribed = df[df['y'] == 'no']['age']

# Perform the independent sample t-test
t_statistic, p_value = ttest_ind(subscribed, not_subscribed)

# Set the significance level
alpha = 0.05

# Make a conclusion based on the p-value
if p_value < alpha:
    print(f"Reject the null hypothesis. {alternative_hypothesis}")
else:
    print(f"Fail to reject the null hypothesis. {null_hypothesis}")
```

```python
# Marketing predictive model to analyse if client has subscribed to term deposit

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

df = pd.read_csv('/content/bank-marketing-data-set/bank-direct-marketing-campaigns.csv', delimiter=',', encoding='Latin-1')

X = df.drop('age', axis=1)
y = df['age']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```
